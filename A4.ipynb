# Research Questions:
Creating the Flavor:
- What ingredients and flavor combinations are most strongly correlated with high ratings in ice cream flavors?
- What commonalities can be found in the ingredients of top-rated flavors from each brand?
- Are positive sentiments in reviews more focused on taste, texture, uniqueness, or other factors?
Creating the Description:
- What commonalities can be found in the flavor descriptions of top-rated flavors from each brand?
- What language or sentiment in customer reviews correlates with higher ratings?

# Hypothesis:
- Flavors with chocolate might be more popular.
- Customers might be inclined towards flavors with interesting mix-ins (cookie dough, chocolate chips, swirls, etc...)
- Flavor descriptions with words such as "Creamy", "Rich", or other indulgent words might rank higher.

# Data selected for analysis:
The data I've selected for this project is from Kaggle. It is a data set that provides insight about various products (ice creams) from Ben & Jerry's, Häagen-Dazs, Breyers, and Talenti. These insights include the flavor names, descriptions, ingredients, ratings, and more. Furthermore, the dataset also provides customer reviews of these ice creams with comments they had on the flavor. The data was collected from the brand websites.

# Methods
First I will filter my data to focus on top-rated ice creams those with an average rating above 4.5/5 stars. From there I will check the data for duplicates, handle NA values, and preprocess text data to work better with my analysis (lowercaing everything, removing punctuation, etc...). 

For figuring out the flavor itself, I plan to use the ingredient lists from the data and identify the most commonly used ingredients across top-rated flavors from each brand. From this, I will create bar-chart visualizations to identify these trends and draw comparisons between brands. This will reveal the building blocks of popular flavors, helping to define the base flavor profile. Furthermore, another option would be to find correlations between base flavors and mix-ins that are popular using a co-occurrence analysis to determine which combinations appear most frequently in high-rated flavors. This could highlight pairings like chocolate-and-fruit or vanilla-and-nut— understanding how certain ingredients pair can help design complementary and well-balanced flavors. Ultimately, I will holistically review all of these visualizations to find key trends and use them to influence my ingredient list for "The Perfect Ice Cream."

For figuring out the description for this perfect ice cream, I will then go back and review all the descriptions for the top-rated flavors with the ingredients from my perfect ice cream flavor. Additionally, I may go through customer sentiments for these flavors and use sentiment analysis to determine what common words were associated with positive reviews. I will find the most common words and create a word tree to use as a vocabulary box while drafting a new description of my flavor. Ultimately, positive language plays a big role in marketing and it can influence sensory and emotional responses among ice-cream lovers.
